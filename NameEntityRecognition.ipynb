```python
sentence="The Eiffel Tower was built from 1887 to 1889 by French engineer Gustave Eiffe whose company specialized in building metal frameworks and structures."
```

```python
import nltk



words=nltk.word_tokenize(sentence)
```

```python
nltk.download('maxent_ne_chunker')
```

```python
nltk.download('words')
```

```python
nltk.download('averaged_perceptron_tagger_eng')
```

```python
nltk.download('maxent_ne_chunker_tab')
```

```python
tag_elements=nltk.pos_tag(words)
```

```python
import matplotlib.pyplot as plt
from nltk.tree import Tree

chunked = nltk.ne_chunk(tag_elements)
plt.figure(figsize=(10,8))
plt.title("NER Tree")
plt.text(0.01, 0.5, chunked.pformat(), fontsize=12)
plt.axis('off')
plt.show()
```
