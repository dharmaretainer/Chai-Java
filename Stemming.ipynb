```python
##Classification Problem
##Comments of product is a positive review or negative review
## Reviews---->eating , eat, eaten-->eat
words=["eating","eats","eaten","writing","writes","programming","program","history","finally","finalized"]
```

```python
###porter stemmer
from nltk.stem import PorterStemmer
stemming=PorterStemmer()
for word in words:
  print(word+"-->"+stemming.stem(word))
```

```python
##RegexpStemmer class
from nltk.stem import RegexpStemmer
reg_stemmer=RegexpStemmer('ing$|s$|e$|able$',min=4)
reg_stemmer.stem('eating')
```

```python
reg_stemmer.stem('imgEating')
```

```python
### Snowball Stemmer
from nltk.stem import SnowballStemmer
snowballStemmer=SnowballStemmer('english')
for word in words:
  print(word+"--->"+snowballStemmer.stem(word))
```

```python
#difference between PorterStemmer and Snowball
stemming.stem("fairly"),stemming.stem("sportingly")
```

```python
#snowball does better
snowballStemmer.stem("fairly"),snowballStemmer.stem("sportingly")
```
